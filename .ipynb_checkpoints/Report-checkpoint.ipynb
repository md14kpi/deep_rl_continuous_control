### 2. Examine the State and Action Spaces

In this environment, a double-jointed arm can move to target locations. A reward of `+0.1` is provided for each step that the agent's hand is in the goal location. Thus, the goal of your agent is to maintain its position at the target location for as many time steps as possible.

The observation space consists of `33` variables corresponding to position, rotation, velocity, and angular velocities of the arm.  Each action is a vector with four numbers, corresponding to torque applicable to two joints.  Every entry in the action vector must be a number between `-1` and `1`.

Run the code cell below to print some information about the environment.

````python
# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents
num_agents = len(env_info.agents)
print('Number of agents:', num_agents)

# size of each action
action_size = brain.vector_action_space_size
print('Size of each action:', action_size)

# examine the state space 
states = env_info.vector_observations
state_size = states.shape[1]
print('There are {} agents. Each observes a state with length: {}'.format(states.shape[0], state_size))
print('The state for the first agent looks like:', states[0])
````

# REPORT

1. Agent Algorithm/Architecture/Hyperparameters
    * LEARNING ALGORITHM: DDPG
    * BUFFER_SIZE = int(1e5)  # replay buffer size, 
    * BATCH_SIZE = 128 # to match input
    * GAMMA = 0.99            # discount factor
    * TAU = 1e-3              # for soft update of target parameters
    * LR_ACTOR = 1e-4         # learning rate of the actor 
    * LR_CRITIC = 1e-4        # learning rate of the critic
    * WEIGHT_DECAY = 0        # L2 weight decay

    * Actor Model Architecture
        * 2 fully-connected dense layers with 128 units
        * 1st layer has relu activation and the output layer has tanh activation

    * Critic Model Architecture
         * 3 full-connected dense layers with 128 units and leaky relu activation
         * Dropout layer applied after third dense layer and before final output dense layer with 1 unit


2. Experimentation Process: 
    * I largely utilized the actor-critic network structure from the Continuous Control from Deep Reinforcement Learning paper, https://arxiv.org/abs/1509.02971 with some tuning of hyperparams
    * Actor and critic dense layer units and batch sizes: I tried a variety of combinations with increasing the units of the initial dense layer to match the batch size and then decreasing it (512, 256) or (256, 128). I found that the learning was unstable for these larger batch and unit sizes and could see this in the environment where the agent would repeat a motion continuously without learning much. Eventually, I found some learning     
    * Learning rates of actor and critic: I tried the learning rates from the DDPG paper where the critic had a higher learning rate than the actor and saw moderate success. However, I found eventually that making them equal with the same network architecture provided the maximal increase in learning.
    * Increasing the number of steps in each episode: After inspecting the trained agent, I realized that the agent had learned an optimal policy but that there were not enough time steps to achieve the target reward. So, I increased the time steps from 300/episode to 700/episode and then finally reaching the goal with 1000/episode. In the future, the project should have a constraint or some set parameters for the maximal number of steps/episode or the maximum number of total steps.
    * Number of episodes needed to solve the environment: 500 episodes (500K time steps)

3. Ideas for Future Work
    * Add batch normalization to actor networks
    * Add priority experience replay to buffer
    * Evaluate PPO and A3C approaches
    * Test same hyperparameters on multi-agent environment


````python
# function that invokes Agent class and trains it with model and agent hyperparameters
def single_agent_ddpg(env, brain_name, update_interval, random_seed,  n_episodes=500, max_t=1000, print_every=100):

    brain = env.brains[brain_name]
    action_size = brain.vector_action_space_size
    scores_deque = deque(maxlen=print_every)
    scores_output = []
    agent = Agent(state_size=33, action_size=action_size, random_seed=random_seed)

    for i_episode in range(1, n_episodes+1):

        env_info = env.reset(train_mode=True)[brain_name]
        # examine the state space 
        states = env_info.vector_observations[0]
        agent.reset()

        scores = 0 
        step_counter = 0
        for t in range(max_t):
            
            actions = agent.act(states, add_noise=False)
            env_info = env.step(actions)[brain_name]        # send the action to the environment
            next_states = env_info.vector_observations[0]   # get the next state
            rewards = env_info.rewards[0]                   # get the reward
            dones = env_info.local_done[0]
            if step_counter < update_interval:
                step_counter += 1            
                if step_counter == update_interval:
                    agent.step(states, actions, rewards, next_states, dones, skip_update=False)
                    step_counter = 0
                    
                else:
                    agent.step(states, actions, rewards, next_states, dones)
            
            scores += rewards

            states = next_states
            

            if np.any(dones):
                break 
                
        print('\rTotal score (averaged over agents) this episode: {}'.format(scores ))
        scores_deque.append(scores)
        scores_output.append(scores)
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_deque)), end="")
        
        torch.save(agent.actor_local.state_dict(), 'checkpoint_actor.pth')
        torch.save(agent.critic_local.state_dict(), 'checkpoint_critic.pth')
        if i_episode % 100 == 0:
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_deque)))
            
    return scores_output
````

````python
# Train the agent
random_seed = 1020
update_interval = 1
single_agent_scores = single_agent_ddpg(env, brain_name, update_interval, random_seed)
````

# After 500 episodes (500K steps) agent achieves objective of average score of 30 over 100 episodes.

````python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(1, len(single_agent_scores)+1), single_agent_scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
````
